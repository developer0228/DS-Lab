```python
import socket, threading, datetime, time

client_data = {}

def receive(conn, addr):
    while True:
        try:
            conn.recv(1024)
            client_data[addr] = conn
            print("Client Data updated with:", addr)
            time.sleep(5)
        except:
            break

def connect(server):
    while True:
        conn, a = server.accept()
        addr = str(a[0]) + ":" + str(a[1])
        print(addr + " got connected successfully")
        threading.Thread(target=receive, args=(conn, addr)).start()

def sync():
    while True:
        print("New synchronization cycle started.")
        print("Number of clients to be synchronized:", len(client_data))

        if client_data:
            for addr, conn in client_data.items():
                try:
                    conn.send(str(datetime.datetime.now()).encode())
                except:
                    pass
        else:
            print("No client data. Synchronization not applicable.")

        print("\n")
        time.sleep(5)

def server_start(port=8080):
    s = socket.socket()
    s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

    print("Socket at master node created successfully\n")

    s.bind(('', port))
    s.listen(10)

    print("Clock server started...\n")
    print("Starting to make connections...\n")
    print("Starting synchronization parallelly...\n")

    threading.Thread(target=connect, args=(s,)).start()
    threading.Thread(target=sync).start()

server_start()
```
